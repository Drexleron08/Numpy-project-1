```python
import numpy as np
climate_data=np.array([[25.0,76.0,99.0],[45.0,68.0,56.0],
                       [12.0,67.0,89.0],[45.0,23.0,78.0],
                       [67.0,34.0,23.0],[78.0,56.0,12.0],
                       [34.0,23.0,45.0],[23.0,45.0,67.0],
                       [56.0,78.0,89.0],[90.0,34.0,23.0],
                       [45.0,67.0,89.0],[23.0,45.0,67.0],
                       [36.0,78.0,90.0],[12.0,34.0,56.0]])
weights=np.array([0.2,0.5,0.3])
average_yeild=np.dot(climate_data,weights)
print("Weighted averages for each region:",average_yeild)

```

```python
print(climate_data.shape)
```

```python
yeild=average_yeild.reshape(14,1)
```

```python
climate_result=np.concatenate((climate_data,yeild),axis=1)
print("Climate data with weighted averages:\n",climate_result)
```
